```
def create_border_array_numpy(rows, cols):
    # Create an array of all ones
    arr_7 = np.ones((rows, cols), dtype=int)
    # Set the inner elements to zero using slicing (from index 1 to second-last)
    if rows > 2 and cols > 2:
        arr_7[1:-1, 1:-1] = 0
    return arr_7

# Example for a 5x5 array
array_5x5 = create_border_array_numpy(5, 5)
print(array_5x5)
```
